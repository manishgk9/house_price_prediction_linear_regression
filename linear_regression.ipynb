```python
# importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('HousingData.csv')
df.head()
```

```python
text="CRIM	ZN	INDUS	CHAS	NOX	RM	AGE	DIS	RAD	TAX	PTRATIO	BLACK	LSTAT	MEDV"
colums=text.split()
print(text)
i=0
for col in colums:
    colums[i]=col.lower()
    i+=1
print(colums)
```

```python
df.columns=colums
```
